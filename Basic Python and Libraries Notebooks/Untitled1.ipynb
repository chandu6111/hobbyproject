```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
import random
x=np.arange(0,10)
y=np.array(random.sample(range(1,50),10))
```

```python
x
```

```python
y
```

# bar plots

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(x,y)
```

```python
plt.bar(x,y,width=0.5)
```

# histogram

```python
dat1=np.array(random.sample(range(1,100),50))
```

```python
dat1
```

```python
fig,ax=plt.subplots()
ax.hist(dat1)
plt.show()
```

```python
plt.hist(dat1,bins=5)
```

```python
groups=["A","B","c"]
stu=[20,30,25]
fig,ax=plt.subplots()
_=ax.pie(stu,labels=groups,autopct="%1.2f%%")
```

```python
x=np.arange(0,50)
fig,ax=plt.subplots()
ax.scatter(x,x,c="b")
ax.scatter(x,dat1,color="g")
```

# box plots

```python
data2=[]
data2.extend(dat1)
data2.extend([-10,-49,-150,150,200,155])
```

```python
data2
```

```python
plt.boxplot(data2)
```

```python
#circles are outliers
#limiting boundaries are max and min values called whiskers
#orange line is median
#in box,from bottom 1st line is 25 percentile, second line is 50 percentile and third is 75 percentile
```
